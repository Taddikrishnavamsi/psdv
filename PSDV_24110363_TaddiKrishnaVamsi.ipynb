3.Fibonacci Sequence:
  Write a Python function that takes a number n and returns a list of the first n Fibonacci numbers.
  Plot the Fibonacci sequence using a line graph.

```python
fab_l=[]
n=int(input("enter number here:"))
def f(n):
    if n<=0:
        return 0
    elif n==1:
        return 1
    return f(n-1)+f(n-2)
f(n)
for i in range(0,n+1):
    fab_l.append(f(i))
print(fab_l)
y=fab_l
import matplotlib.pyplot as plt
plt.plot(y)

plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line graph')

plt.show()
```

4.List Manipulation:
  Given a list [3, 5, 7, 9, 11, 13]:
  Insert 6 at the 3rd index.
  Remove the number 7 from the list.
  Reverse the list.
  Calculate the mean and median of the modified list.

```python
l = [3, 5, 7, 9, 11, 13]

l.insert(3,6)
print(l)
l.remove(7)
print(l)
l = l[::-1]
print(l)
sum = 0
for i in l:
    sum += i
mean = sum / len(l)
median = (l[2]+l[3])/2
print(f"median:{median}")
print(f"mean:{mean:.2f}")
```

9.Merging Dictionaries:
  Write Python code to merge two dictionaries. If a key is present in both, sum their values.

  Example:

  dict1 = {'a': 10, 'b': 20}
  dict2 = {'b': 5, 'c': 15}
  Output: {'a': 10, 'b': 25, 'c': 15}

```python
dict1 = {'a': 10, 'b': 20}
dict2 = {'b': 5, 'c': 15}
dict_merge={}
for keys in dict1:
    if keys in dict2:
        dict_merge[keys]=dict1[keys]+dict2[keys]
    else:
        dict_merge[keys]=dict1[keys]
for keys in dict2:
    if keys not in dict1:
        dict_merge[keys]=dict2[keys]
print(dict_merge)
```

13.Sorting Dictionaries:
   Write a function to sort a list of dictionaries by a key.

   Example:

   data = [{'name': 'Alice', 'age': 25}, {'name': 'Bob', 'age': 22}]
   Sort by age → [{'name': 'Bob', 'age': 22}, {'name': 'Alice', 'age': 25}]

```python
data = [{'name': 'Alice', 'age': 25}, {'name': 'Bob', 'age': 22}]
for i in range(len(data)-1):
    for j in range(0,len(data)-i-1):
       if data[j]["age"]>=data[j+1]["age"]:
          data[j],data[i+1]=data[j+1],data[i]
print(f"sort by age:{data}")
```

15.Anagram Finder:
   Write a Python function to find all anagrams of a word in a given list.
   Input: word = "listen", words = ["enlist", "google", "inlets", "banana"]
   Output: ["enlist", "inlets"]

```python
words = ["enlist", "google", "inlets", "banana"]
list=[]
words_new=[]
for char in words:
    char=sorted(char)
    words_new.append(char)
for i in range(len(words_new)):
    for j in range(i+1,len(words_new)):
        if words_new[i]==words_new[j]:
            list.append(words[i])
            list.append(words[j])
print(list)
```

22.Normalization:
Write a Python function to normalize a list of numbers to a range of 0 to 1.
Visualize the normalized data using a line plot.


```python
import matplotlib.pyplot as plt
lis = [5, 15, 10, 25, 20, 30, 35]
nor_lis=[]
minval=min(lis)
maxval=max(lis)
for x in lis:
    nor_val=(x-minval)/(maxval-minval)
    nor_lis.append(nor_val)

print(lis)
print(nor_lis)

```
